```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error , r2_score
```

```python
data = pd.read_csv("../Datasets/housing.csv")
data = data.drop("ID" , axis = 1)
data.head()
```

```python
data.isnull().sum()
```

```python
X = data.drop('medv' , axis = 1)
y = data['medv']
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20 , 20))
sns.heatmap(data.corr() , annot = True)
```

```python
sns.regplot(data = data , x = data['lstat'] , y = 'medv')
```

```python
sns.regplot(data = data , x = 'chas' , y = 'medv')
```

```python
# so we can drop chas , dist to make the model more accurate and decrease its dependency
# with this variable
# not done here but you only need to drop columns
```

```python
X_train , X_test , y_train , y_test = train_test_split(X , y , test_size=0.8  , random_state = 42)
```

```python
model = LinearRegression()

model.fit(X_train , y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred.)
```

```python
MSE = mean_squared_error(y_test , y_pred)
r2 = r2_score(y_pred , y_test)
```

```python
print("mean squared Error : ", MSE)
print("r2 Score : " , r2)
```
